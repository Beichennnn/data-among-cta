```python
# subsection of my DateFrame - I want a new DataFrame where I only use active complaints. 
# think of it like a filter
# list comprenhensions! 2 dimensions - 2 sets of brackets
active_complaints_df = complaints_df[[complaint for complaint in complaints_df.Status == 'ACTIVE']]
```

```python
elev_accidents_df = complaints_df[[complaint for complaint in complaints_df['Complaint Category'] == '81']]
```

```python
display(elev_accidents_df)
```

```python
# tolist() can convert a pandas list into a python list
street_names = elev_accidents_df['House Street'].tolist()
# overwrite my variable to get rid of white space
# street_names = [street.rstrip() for street in street_names]

house_numbers = elev_accidents_df['House Number'].tolist()
print(house_numbers)
```

```python
# put together, using list comprehension to merge two lists together
bad_places = [ num + " " + street for num, street in zip(house_numbers, street_names)]
print(bad_places)
```

```python
# there are the places to NEVER take the elevator
from collections import Counter
# give me the top 10 most reported addresses with elevator accidents
worst_places = Counter(bad_places).most_common(10)

print(worst_places)
```
